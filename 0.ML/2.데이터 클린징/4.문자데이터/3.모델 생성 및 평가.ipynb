```python
import pandas as pd
df=pd.read_csv('data/loan_train_cleaning.csv')
df.head().T
```

### 머신러닝 변수
* label : 종속 변수 또는 정답
* feature : 독립변수 또는 속성

```python
df.columns
```

```python
features = ['Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area']
label = 'Loan_Status'
X,y = df[features], df[label] #x는 과정, y는 정답
```

```python
X.T
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = \
train_test_split(X,y, test_size=0.2)#확인용 테스트는 20% , 트레인에 80%
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

### 모델 생성 및 학습
* 모델 생성 : 변수 = DecisionTreeClassifier()
* 학습 : 변수.fit(X_train, y_train)
    - 문제와 정답을 알고리즘에 학습
* 예측 : 변수.predict(X_test)
    - 확인용 문제를 predict에 넣으면 정답을 알려준다

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier() #모델 생성
dtc.fit(X_train, y_train)# 학습
```

```python
pred = dtc.predict(X_test) #예측된 결과'pred'
pred #예측된 정답
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score #정확도를 ㅂ여줌.
```

```python
accuracy_score(y_test, pred) #얼마나 일치하냐. 
```
